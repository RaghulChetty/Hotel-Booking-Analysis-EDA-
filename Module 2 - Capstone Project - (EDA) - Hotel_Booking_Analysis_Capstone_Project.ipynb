```python
# Checking Outliers for continous variables
from scipy.stats import norm
for i in continous_variables :
  plt.figure(figsize = (15,7))
  plt.subplot(1,2,1)
  ax = sns.boxplot(data = df[i])
  ax.set_title(f'{i}')
  ax.set_ylabel(i)

  plt.subplot(1,2,2)
  ax = sns.distplot(df[i], fit = norm)
  ax.set_title(f'skewness of {i} : {df[i].skew()}')
  ax.set_xlabel(i)
  print('\n')
  plt.show()
```

  Outliers were found in the variables lead_time and adr, but not in the variable agent.

```python
# Removing Outliers using Inter Quantile Range (IQR) in skew symmetric curve

# Outlier columns
outlier_columns = ['lead_time','adr']

# copy dataset as new_df
new_df = df.copy()

# Capping dataset
for i in outlier_columns :
  # Find IQR
  Q1 = new_df[i].quantile(0.25)
  Q3 = new_df[i].quantile(0.75)
  IQR = Q3-Q1

  # define lower and upper limit
  lower_limit = new_df[i].quantile(0.25)-1.5*IQR
  upper_limit = new_df[i].quantile(0.75)+1.5*IQR

  # apply lower and upper limit to each variables
  new_df.loc[(new_df[i] > upper_limit),i] = upper_limit
  new_df.loc[(new_df[i] < lower_limit),i] = lower_limit
```

```python
# Checking Outliers for continous variables
from scipy.stats import norm
for i in outlier_columns :
  plt.figure(figsize = (15,7))
  plt.subplot(1,2,1)
  ax = sns.boxplot(data = new_df[i])
  ax.set_title(f'{i}')
  ax.set_ylabel(i)

  plt.subplot(1,2,2)
  ax = sns.distplot(new_df[i], fit = norm)
  ax.set_title(f'skewness of {i} : {new_df[i].skew()}')
  ax.set_xlabel(i)
  print('\n')
  plt.show()
```

```python
# changing Datatype of Children and Agent variables from float64 to int64
new_df[['children','agent']] = new_df[['children','agent']].astype('int64')

# changing Datatype of Reservation_status_date variable from object to datetime64
new_df['reservation_status_date'] = pd.to_datetime(new_df['reservation_status_date'],format = '%Y-%m-%d')
```

```python
# check datatype
new_df[['children','agent','reservation_status_date']].info()
```

Variables children and agent are in float64 format, so converted them to the appropriate datatype, int64. The reservation_status_date variable is in object format, so converted it to datetime64 format.

```python
# ADD Night stays on week night and weekend night in one variable 'total_stays'
new_df['total_stays'] = new_df['stays_in_week_nights'] + new_df['stays_in_weekend_nights']

# ADD Adults,Children,Babies to variable 'total_people'
new_df['total_people']= new_df['adults']+ df['children']+ new_df['babies']

# ADD children and Babies to variable 'total_childrens'
new_df['total_childrens']= df['children']+ new_df['babies']

# Create 'guest_category' from variable 'total_people'
new_df['guest_category']=np.where(new_df['total_people']==1, 'single',
                                 np.where(new_df['total_people']==2, 'couple', 'family'))

# Create 'reserved_room_assigned' variable which describes same Room assigned or not
new_df['reserved_room_assigned'] = np.where(new_df['reserved_room_type'] == new_df['assigned_room_type'],'yes','no')

# Create 'lead_time_category' from 'lead_time' variable to display category
new_df['lead_time_category'] = np.where(new_df['lead_time']<=15, 'low',
                                 np.where((new_df['lead_time']>15) & (new_df['lead_time']<90), 'medium', 'high'))


# Check Dataset
new_df.head()
```

```python
# Remove observations having value 0 in 'total_people' variable
new_df.drop(new_df[new_df['total_people'] == 0].index, inplace = True)
```

**Insights**

- For more ease, "total_stays" variable is created, by adding the variables "stays_in_weekend_nights" and "stays_in_weeknights" the variable "total_stays" is created. And "total_people" is created, by combining the variables "adults," "children," and "babies".
- The variable "reserved_room_assigned" is made up of the variables "reserved_room_type" and "assigned_room_type", which describes whether  reserved room for a customer has been assigned or not.
- From variables "children" and "babies," a new "total_children" variable is created by adding both of them.
- The variable "total_people" was used to create "guest_category," which describes bookings made for individuals, couples, or families.
- The variable 'lead_time_category' was created from the variable 'lead_time' to display lead time as low, medium, and high.
- Variable "total_people" cannot be zero, so observations in total_people having 0 values are dropped.

```python
# Obtain target variable
excluded_variables = [var for var in new_df.columns if len(new_df[var].value_counts()) > 15]
target_variables = [var for var in new_df.columns if var not in excluded_variables]

# define 'r' to autofit the number and size of plots
r = int(len(target_variables)/3 +1)
```

```python
# define a function to Notate the percentage count on each value on the bars
def annot_percent(axes) :
  "Takes axes as input and labels the percentage count of each bar in countplot"
  for p in plot.patches :
    total = sum(p.get_height() for p in plot.patches)/100
    percent = round(p.get_height()/total,2)
    x  = p.get_x() + p.get_width()/2
    y = p.get_height()
    plot.annotate(f'{percent}%',(x,y), ha='center', va='bottom')
```

```python
# plotting countplots for each variable in target_variables
plt.figure(figsize = (18,r*3))
for n,var in enumerate(target_variables):
  plot = plt.subplot(r,3,n+1)
  sns.countplot(x = new_df[var]).margins(y=0.15)
  plt.title(f'{var.title()}',weight='bold')
  plt.tight_layout()
  annot_percent(plot)
```

Because customers do not prefer to book room types B, C, H, and L, the hotel can eliminate them, lowering the cost of these rooms.

- Because 3 and 8 parking spaces were rarely requested by customers, hotels can only keep bookings for 1 and 2 parking spaces to save money.
- 15% of customers were not given reserved rooms. Make sure that guests get the rooms they have booked.
- Almost 25% of customers cancelled their bookings. Hotel should implement a cancellation policy, discount on confirmed bookings, and send booking reminders to guests to reduce booking cancellations.
